```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
d = pd.read_csv('Salary_dataset - Salary_dataset.csv')
d.head(8)
```

```python
plt.title('Salary vs Experience')
plt.xlabel('YearsExperience (Years)')
plt.ylabel('Salary (PAK PKR)')
plt.scatter(d.YearsExperience,d.Salary, color='red', marker= '.')
```

```python
reg = linear_model.LinearRegression()
reg.fit(d[['YearsExperience']], d.Salary)
```

```python
# Predicting Salary for a given YearsExperience
years_of_experience = 6.5  # Example value, change as needed
predicted_salary = reg.predict([[years_of_experience]])
print(f"Predicted salary for {years_of_experience} years of experience: {predicted_salary[0]}")

# Visualizing the Regression Line
plt.plot(d.YearsExperience, reg.predict(d[['YearsExperience']]), color='blue')
plt.scatter(years_of_experience, predicted_salary, color='green', marker='o', s=100)

plt.title('Salary vs Experience')
plt.xlabel('YearsExperience (Years)')
plt.ylabel('Salary (PAK PKR)')

plt.show()

# Getting the coefficients (slope and intercept)
print("Coefficients:", reg.coef_)
print("Intercept:", reg.intercept_)
```
